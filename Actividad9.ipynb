## Ejercicio 1

$$\begin{eqnarray*}
\frac{d x}{dt} & = & y  \\
\frac{d y}{dt} & = & -x   \\
\end{eqnarray*}$$

$$\begin{eqnarray*}
\frac{d\vec{x}}{dt} & = & \begin{bmatrix}
0 & 1 \\ -1 & 0 \end{bmatrix} \begin{bmatrix}
x \\ y \end{bmatrix}  \\
\end{eqnarray*}$$

```python
# Definimos a, b, c, d:
a = 0
b = 1
c = -1
d = 0

# Definimos a la matriz A:
A = np.array([[a, b],[c, d]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
print('Los eigenvectores son:')
print(eigenVektor)
print('Los eigenvalores son:', eigenWert)

# Calculamos la traza:
p = eigenWert[0] + eigenWert[1]
q = eigenWert[0]*eigenWert[1]

print('La traza es: Tr(A) = ', p)

# Calculamos el discriminante:
print('El discriminante es: Δ = ', p**2 - 4*q)
```

El punto crítico es: *Centro*


Para encontrar la solución derivamos la primera ecuación:

$$ \frac{d^2 x}{dt^2} = \frac{dy}{dt} = -x $$

$$\frac{d^2 x}{dt^2} + x = 0$$

Resolvemos polinomio característico:

$$ m^2 + 1 = 0 $$
$$ m = \pm i $$



La solución de $x(t)$ es:

$$ x(t) = c_1 \cos{t} + c_2 \sin{t}   $$ 

Derivamos $x(t)$:

$$ \frac{dx}{dt} = -c_1 \sin{t} + c_2 \cos{t} $$

Por la segunda ecuación diferencial:

$$ \frac{dx}{dt} = y $$

La solución del sistema de ecuaciones es:

$$ x(t) = c_1 \cos{t} + c_2 \sin{t}   $$ 
$$ y(t) = -c_1 \sin{t} + c_2 \cos{t} $$

```python
# Definimos la función solución:
def S(c1, c2, t):
  x1 = c1*np.cos(t) + c2*np.sin(t)
  x2 = c2*np.cos(t) - c1*np.sin(t)
  x = [x1, x2]
  return np.array(x)

# Generamos distintas constantes

t = np.linspace(-25, 25, 5000)
c1 = np.linspace(-25, 25, 11)
c2 = np.linspace(20, 70, 11)

# Graficamos las soluciones

plt.figure(figsize=(8, 8))
for i, j in zip(c1, c2):
  sol = np.array(S(i,j,t))
  plt.plot(sol[0,:],sol[1,:])
plt.title('Soluciones particulares del sistema de ecuaciones')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
```

```python
# Definimos a, b, c, d:
a = 0
b = 1
c = 1
d = 0

# Definimos a la matriz A:
A = np.array([[a, b],[c, d]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
print('Los eigenvectores son:')
print(eigenVektor)
print('Los eigenvalores son:', eigenWert)

# Calculamos la traza:
p = eigenWert[0] + eigenWert[1]
q = eigenWert[0]*eigenWert[1]

print('La traza es: Tr(A) = ', p)

# Calculamos el discriminante:
print('El discriminante es: Δ = ', p**2 - 4*q)
```

El punto crítico es: *Punto Silla*

Los eigenvectores son:

$$ v_1 = \begin{pmatrix} 1 \\ 1 \end{pmatrix} $$
$$ v_2 = \begin{pmatrix} -1 \\ 1 \end{pmatrix} $$

La solución del sistema de ecuaciones es:

$$ x(t) = c_1 e^t - c_2 e^{-t}   $$ 
$$ y(t) = c_1 e^t + c_2 e^{-t} $$

```python
# Definimos la función solución:
def S(c1, c2, t):
  x1 = c1*np.exp(t) - c2*np.exp(-t)
  x2 = c1*np.exp(t) + c2*np.exp(-t)
  x = [x1, x2]
  return np.array(x)

# Generamos distintas constantes

t = np.linspace(-3, 3, 5000)
c1 = [5, 5, 10, 10, -5, -5, -10, -10]
c2 = [5, -5, 10, -10, -5, 5, -10, 10]

# Graficamos las soluciones

plt.figure(figsize=(8, 8))
for i, j in zip(c1, c2):
  sol = np.array(S(i,j,t))
  plt.plot(sol[0,:],sol[1,:])
plt.title('Soluciones particulares del sistema de ecuaciones')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
```

## Ejercicio 3

\begin{eqnarray*}
\frac{d^2 x}{d t^2} + \omega_0^2 x & = & 0, \ \omega_0 > 0  \\
\end{eqnarray*}

$$ \frac{dx}{dt} = v $$
$$ \frac{dv}{dt} = -\omega_0^2x$$

$$\begin{eqnarray*}
\frac{d\vec{x}}{dt} & = & \begin{bmatrix}
0 & 1 \\ -\omega_0^2 & 0 \end{bmatrix} \begin{bmatrix}
x \\ v \end{bmatrix}  \\
\end{eqnarray*}$$

```python
# Elegimos un valor para omega_0:
w0 = 3

# Definimos a, b, c, d:
a = 0
b = 1
c = -w0**2
d = 0

# Definimos a la matriz A:
A = np.array([[a, b],[c, d]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
print('Los eigenvectores son:')
print(eigenVektor)
print('Los eigenvalores son:', eigenWert)

# Calculamos la traza:
p = eigenWert[0] + eigenWert[1]
q = eigenWert[0]*eigenWert[1]

print('La traza es: Tr(A) = ', p)

# Calculamos el discriminante:
print('El discriminante es: Δ = ', p**2 - 4*q)
```

El punto crítico es: *Centro*


Para encontrar la solución derivamos la primera ecuación:

$$ \frac{d^2 x}{dt^2} = \frac{dv}{dt} = -\omega_0^2x $$

$$\frac{d^2 x}{dt^2} + \omega_0^2x = 0$$

Resolvemos polinomio característico:

$$ m^2 + \omega_0^2 = 0 $$
$$ m = \pm \omega_0 i $$



La solución de $x(t)$ es:

$$ x(t) = c_1 \cos{\omega_0 t} + c_2 \sin{\omega_0 t} $$ 

Derivamos $x(t)$:

$$ \frac{dx}{dt} = -c_1 \omega_0 \sin{\omega_0 t} + c_2 \omega_0 \cos{\omega_0 t} $$

Por la segunda ecuación diferencial:

$$ \frac{dx}{dt} = v $$

La solución del sistema de ecuaciones es:

$$ x(t) = c_1 \cos{\omega_0 t} + c_2 \sin{\omega_0 t} $$ 
$$ v(t) = -c_1 \omega_0 \sin{\omega_0 t} + c_2 \omega_0 \cos{\omega_0 t} $$

```python
# Definimos la función solución:
def S(c1, c2, t):
  x1 = c1*np.cos(w0*t) + c2*np.sin(w0*t)
  x2 = c2*w0*np.cos(w0*t) - c1*w0*np.sin(w0*t)
  x = [x1, x2]
  return np.array(x)

# Generamos distintas constantes

t = np.linspace(-25, 25, 5000)
c1 = np.linspace(-25, 25, 11)
c2 = np.linspace(20, 70, 11)

# Graficamos las soluciones

plt.figure(figsize=(10,10))
for i, j in zip(c1, c2):
  sol = np.array(S(i,j,t))
  plt.plot(sol[0,:],sol[1,:])
plt.title('Soluciones particulares del sistema de ecuaciones')
plt.xlim(-200, 200)
plt.ylim(-200, 200)
plt.xlabel('x(t)')
plt.ylabel('y(t)')
```

## Ejercicio 4

\begin{eqnarray*}
\frac{d x}{d t} & = & -2 x  \\
\frac{d y}{dt } & = &  2 z  \\
\frac{d z}{dt } & = &  - 2 y  \\
\end{eqnarray*}

$$\begin{eqnarray*}
\frac{d\vec{p}}{dt} & = & \begin{bmatrix}
0 & 2 \\ -2 & 0 \end{bmatrix} \begin{bmatrix}
y \\ z \end{bmatrix}  \\
\end{eqnarray*}$$

```python
# Definimos a la matriz A:
A = np.array([[0, 2],[-2, 0]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
print('Los eigenvectores son:')
print(eigenVektor)
print('Los eigenvalores son:', eigenWert)

# Calculamos la traza:
p = eigenWert[0] + eigenWert[1] 
q = eigenWert[0]*eigenWert[1]

print('La traza es: Tr(A) = ', p)

# Calculamos el discriminante:
print('El discriminante es: Δ = ', p**2 - 4*q)
```

El punto crítico es: *Centro*

Resolvemos la primera ecuación por separación de variables

$$ -\frac{dx}{2x} = dt $$

$$ -\frac{1}{2} \int \frac{dx}{x} = \int dt$$
$$ -\frac{1}{2} \ln{|x|} = t + C $$
$$ x(t) =  c_1 e^{-2t}$$

Para encontrar la solución derivamos la primera ecuación:

$$ \frac{d^2 y}{dt^2} = 2\frac{dz}{dt} = -4y $$

$$\frac{d^2 y}{dt^2} + 4y = 0$$

Resolvemos polinomio característico:

$$ m^2 + 4 = 0 $$

$$ m = \pm 2i $$

La solución de $y(t)$ es:

$$ y(t) = c_2 \cos{2t} + c_3 \sin{2t}   $$ 

Derivamos $y(t)$:

$$ \frac{dy}{dt} = -2 c_2 \sin{2t} + 2 c_3 \cos{2t} $$

Por la segunda ecuación diferencial:

$$ \frac{1}{2} \frac{dy}{dt} = z $$

La solución del sistema de ecuaciones es:

$$ x(t) =  c_1 e^{-2t} $$
$$ y(t) = c_2 \cos{2t} + c_3 \sin{2t} $$ 
$$ z(t) = -c_2 \sin{2t} + c_3 \cos{2t} $$

```python
# Definimos la función solución:
def S(c1, c2, c3, t):
  x1 = c1*np.exp(-2*t)
  x2 = c2*np.cos(2*t) + c3*np.sin(2*t)
  x3 = -c2*np.sin(2*t) + c3*np.cos(2*t)
  x = [x1, x2, x3]
  return np.array(x)

# Generamos distintas constantes

t = np.linspace(-50, 50, 5000)
c1 = np.linspace(-25, 25, 11)
c2 = np.linspace(-25, 25, 11)
c3 = np.linspace(-25, 25, 11)

# Graficamos las soluciones

plt.figure()
ax = plt.axes(projection='3d')
for i, j, k in zip(c1, c2, c3):
  sol = np.array(S(i,j,k,t))
  ax.plot3D(sol[0,:], sol[1,:], sol[2,:])
plt.title('Soluciones particulares del sistema de ecuaciones')
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')
```

## Ejercicio 5

\begin{eqnarray*}
\frac{d x}{d t} & = & - x + z  \\
\frac{d y}{dt } & = &  3 y  \\
\frac{d z}{dt } & = &  - x - z \\
\end{eqnarray*}

$$\begin{eqnarray*}
\frac{d\vec{p}}{dt} & = & \begin{bmatrix}
-1 & 1 \\ -1 & -1 \end{bmatrix} \begin{bmatrix}
x \\ z \end{bmatrix}  \\
\end{eqnarray*}$$

```python
# Definimos a la matriz A:
A = np.array([[-1, 1],[-1, -1]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
print('Los eigenvectores son:')
print(eigenVektor)
print('Los eigenvalores son:', eigenWert)

# Calculamos la traza:
p = eigenWert[0] + eigenWert[1] 
q = eigenWert[0]*eigenWert[1]

print('La traza es: Tr(A) = ', p)

# Calculamos el discriminante:
print('El discriminante es: Δ = ', p**2 - 4*q)
```

El punto crítico es: *Espiral Estable*

Resolvemos la primera ecuación por separación de variables

$$ \frac{dy}{3y} = dt $$

$$ \frac{1}{3} \int \frac{dy}{y} = \int dt$$
$$ \frac{1}{3} \ln{|y|} = t + C $$
$$ y(t) =  c_2 e^{3t}$$

Para encontrar la solución del sistema, despejamos z de la primera ecuación:

$$z = \frac{dx}{dt} + x$$

Derivamos z:

$$\frac{dz}{dt} = \frac{d^2 x}{dt^2} + \frac{dx}{dt}$$

Sustituimos la tercera ecuación:

$$\frac{d^2 x}{dt^2} + \frac{dx}{dt} + x + z = 0$$

Sustituimos z:

$$\frac{d^2 x}{dt^2} + 2\frac{dx}{dt} + 2x = 0$$

Resolvemos polinomio característico:

$$ m^2 + 2m^2 + 2 = 0 $$

$$ m = -1 \pm i $$

La solución de $y(t)$ es:

$$ x(t) = e^{-t} (c_1 \cos{t} + c_3 \sin{t})$$ 

Derivamos $x(t)$:

$$ \frac{dx}{dt} = -e^{-t} (c_1 \cos{t} + c_3 \sin{t}) + e^{-t} (-c_1 \sin{t} + c_3 \cos{t}) = e^{-t} [(c_3 - c_1) \cos{t} - (c_1 + c_3) \sin{t}] $$

Sustituimos para obtener z(t):

$$ z(t) = \frac{dx}{dt} + x $$
$$ z(t) = e^{-t} (c_3 \cos{t} -c_1 \sin{t})$$

La solución del sistema de ecuaciones es:

$$ x(t) = e^{-t} (c_1 \cos{t} + c_3 \sin{t})$$
$$ y(t) = c_2 e^{3t} $$ 
$$ z(t) = e^{-t} (c_3 \cos{t} -c_1 \sin{t}) $$

```python
# Definimos la función solución:
def S(c1, c2, c3, t):
  x1 = np.exp(-t)*(c1*np.cos(t) + c3*np.sin(t))
  x2 = c2*np.exp(3*t)
  x3 = np.exp(-t)*(c3*np.cos(t) - c1*np.sin(t))
  x = [x1, x2, x3]
  return np.array(x)

# Generamos distintas constantes

t = np.linspace(-5, 1.5, 5000)
c1 = np.linspace(-25, 25, 11)
c2 = np.linspace(-25, 25, 11)
c3 = np.linspace(-10, 10, 11)


# Graficamos las soluciones

plt.figure()
ax = plt.axes(projection='3d')
for i, j, k in zip(c1, c2, c3):
  sol = np.array(S(i,j,k,t))
  ax.plot3D(sol[0,:], sol[1,:], sol[2,:])
plt.title('Soluciones particulares del sistema de ecuaciones')
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')
```

## Ejercicio 6

\begin{eqnarray*}
\frac{d x}{d t} & = & - x  \\
\frac{d y}{dt } & = &  x + 2 y  \\
x(0) = 0, & & y(0) = 3
\end{eqnarray*}

$$\begin{eqnarray*}
\frac{d\vec{p}}{dt} & = & \begin{bmatrix}
-1 & 0 \\ 1 & 2 \end{bmatrix} \begin{bmatrix}
x \\ y \end{bmatrix}  \\
\end{eqnarray*}$$

```python
# Definimos a la matriz A:
A = np.array([[-1, 0],[1, 2]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
print('Los eigenvectores son:')
print(eigenVektor)
print('Los eigenvalores son:', eigenWert)

# Calculamos la traza:
p = eigenWert[0] + eigenWert[1] 
q = eigenWert[0]*eigenWert[1]

print('La traza es: Tr(A) = ', p)

# Calculamos el discriminante:
print('El discriminante es: Δ = ', p**2 - 4*q)
```

El punto crítico es: *Punto Silla*

```python
# Definimos la función F(t,p), el lado derecho de la ecuacion diferencial
# La solución p es un arreglo 1-D, p=(x, y), dp/dt es un arreglo

def EDO6(p, t):
        x, y = p
        dpdt = [-x, x+2*y]
        return dpdt

# Definimos las condiciones iniciales:
p0 = [0, 3]
```

```python
# Definimos el tiempo de integración de 0 a 10 segundos 
t = np.linspace(0, 5, 5000)

# Usamos el método de integración de scipy.integrate.odeint
p = odeint(EDO6, p0, t)

plt.figure(figsize=(6,5))
plt.plot(p[:, 0], p[:, 1], 'b', label='x(t) v y(t)')
plt.legend(loc='best')
plt.title('Solución de Sistema de Ecuaciones en el Espacio Fase')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.grid()
plt.show()

plt.figure(figsize=(6,5))
plt.plot(t, p[:, 0], 'b', label='x(t)')
plt.plot(t, p[:, 1], 'g', label='y(t)')
plt.legend(loc='best')
plt.title('Soluciones de x(t) y y(t) con respecto al tiempo')
plt.xlabel('t')
plt.ylabel(' ')
plt.grid()
plt.show()
```

\frac{d x}{d t} & = & A x  \\
x(0) & = & (0,-b,b) \\
A & = & \begin{bmatrix}
2 & 0 & 0 \\
0 & -1 & 0 \\
0 & 2 & -3 \\
\end{bmatrix}
\end{eqnarray*}


$$\begin{eqnarray*}
\frac{d\vec{q}}{dt} & = & \begin{bmatrix}
-1 & 0 \\ 2 & -3 \end{bmatrix} \begin{bmatrix}
x_2 \\ x_3 \end{bmatrix}  \\
\end{eqnarray*}$$

```python
# Definimos a la matriz A:
A = np.array([[-1, 0],[2, -3]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
print('Los eigenvectores son:')
print(eigenVektor)
print('Los eigenvalores son:', eigenWert)

# Calculamos la traza:
p = eigenWert[0] + eigenWert[1] 
q = eigenWert[0]*eigenWert[1]

print('La traza es: Tr(A) = ', p)

# Calculamos el discriminante:
print('El discriminante es: Δ = ', p**2 - 4*q)
```

El punto crítico es: *Nodo Estable*

```python
# Definimos la función F(t,x), el lado derecho de la ecuacion diferencial
# La solución x es un arreglo 1-D, x=(x1, x2, x3), dx/dt es un arreglo

def EDO9(x, t):
        x1, x2, x3 = x
        dxdt = [2*x1, -x2, 2*x2 -3*x3]
        return dxdt

# Definimos las condiciones iniciales:
b = 5
x0 = [0, -b, b]
```

```python
# Definimos el tiempo de integración de 0 a 10 segundos 
t = np.linspace(0, 5, 5000)

# Usamos el método de integración de scipy.integrate.odeint
x = odeint(EDO9, x0, t)


plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x[:, 0], x[:, 1], x[:, 2], 'b', label='x1(t) v x2(t) v x3(t)')
plt.legend(loc='best')
plt.title('Solución de Sistema de Ecuaciones en el Espacio Fase')
ax.set_xlabel('x1(t)')
ax.set_ylabel('x2(t)')
ax.set_zlabel('x3(t)')
plt.show()

plt.figure(figsize=(6,5))
plt.plot(t, x[:, 0], 'b', label='x1(t)')
plt.plot(t, x[:, 1], 'g', label='x2(t)')
plt.plot(t, x[:, 2], 'r', label='x3(t)')
plt.legend(loc='best')
plt.title('Soluciones de x1(t), x2(t), y x3(t) con respecto al tiempo')
plt.xlabel('t')
plt.ylabel(' ')
plt.grid()
plt.show()
```

## Ejercicio 10

Se tiene el siguiente sistema de resortes acoplados con dos masas.  

Screen Shot 2021-03-18 at 00.35.32.png

El sistema está sujeto del primer resorte de la izquierda. Cuando el sistema se deja en reposo, las longitudes de los resortes son $L_1$ y $L_2$.

Las ecuaciones de movimiento estan dadas por las siguientes ecuaciones diferenciales acopladas:

\begin{eqnarray*}
m_1 x_1^{''} + b_1 x_1^{'}+k_1 (x_1-L_1)-k_2(x_2-x_1-L_2) & = & 0 \\
m_2 x_2^{''} + b_2 x_2^{'}+k_2(x_2-x_1-L_2) & = & 0 \\
\end{eqnarray*}

Encuentre las soluciones como funciones de $t$ y grafíquelas, así como las trayectorias en el espacio fase.

Suponga que $L_1=L_2=1$ y que las masas son iguales $m_1=m_2=1$.  

**10.1)** Encuentre los eigenvalores del sistema y diga como son las soluciones.  

**10.2)** Caso sin fricción. $b_1=b_2=0$, $k_1=6$, $k_2=4$, condiciones iniciales $(x_1(0), x_1^{'}(0), x_2(0), x_2^{'}(0))=(1,0,4,0)$ 

**10.3)** Igual que el caso anterior pero, con fricción: $b_1=0.1, b_2=0.2$







```python
# Definimos la función F(t,p), el lado derecho de la ecuacion diferencial
# La solución p es un arreglo 1-D, p=(x1, v1, x2, v2), dp/dt es un arreglo

def EDO10(p, t, k1, k2, b1, b2):
        x1, v1, x2, v2 = p
        dpdt = [v1, -x1*(k1+k2) -b1*v1+k2*x2+k1-k2, v2, k2*x1-k2*x2-b2*v2+k2] 
        return dpdt

# Definimos las condiciones iniciales:
p0 = [1.2, 0, 3.0, 0]
```

Buscamos un sistema de ecuaciones de la forma:
$$ \frac{d \vec{p}}{dt} = {\bf{J}} \vec{p} $$

donde 

$$ \vec{p} = \begin{pmatrix} x_1 \\ v_1 \\ x_2 \\ v_2 \end{pmatrix} $$ \\

$ \vec{p_0} $ es el punto crítico

$$ \vec{p_0} = \begin{pmatrix} \bar{x_1} \\ \bar{v_1} \\ \bar{x_2} \\ \bar{v_2} \end{pmatrix} $$ \\

y ${\bf{J}}$ es la Matriz Jacobiana

$${\bf{J}} = \begin{bmatrix}
\frac{\partial f}{\partial x_1}(\vec{p_0}) & \frac{\partial f}{\partial v_1}(\vec{p_0}) & \frac{\partial f}{\partial x_2}(\vec{p_0}) & \frac{\partial f}{\partial v_2}(\vec{p_0}) \\
\frac{\partial g}{\partial x_1}(\vec{p_0}) & \frac{\partial g}{\partial v_1}(\vec{p_0}) &
\frac{\partial g}{\partial x_2}(\vec{p_0}) & \frac{\partial g}{\partial v_2}(\vec{p_0}) \\
\frac{\partial h}{\partial x_1}(\vec{p_0}) & \frac{\partial h}{\partial v_1}(\vec{p_0}) & \frac{\partial h}{\partial x_2}(\vec{p_0}) & \frac{\partial h}{\partial v_2}(\vec{p_0}) \\
\frac{\partial i}{\partial x_1}(\vec{p_0}) & \frac{\partial i}{\partial v_1}(\vec{p_0}) & \frac{\partial i}{\partial x_2}(\vec{p_0}) & \frac{\partial i}{\partial v_2}(\vec{p_0}) \\
\end{bmatrix} $$ \\

Para encontrar ${\bf{J}}$ requerimos encontrar $ \vec{p_0} $ primero

Sustituimos los valores $m_1 = m_2 = 1$ y $L_1 = L_2 = 1$ en las ecuaciones diferenciales:

\begin{eqnarray*}
x_1^{''} + b_1 x_1^{'}+k_1 (x_1 - 1)-k_2(x_2-x_1-1) & = & 0 \\
x_2^{''} + b_2 x_2^{'}+k_2(x_2-x_1- 1) & = & 0 \\
\end{eqnarray*}

Reescribimos las ecuaciones:

$$ f(v_1) = \frac{dx_1}{dt} = v_1 $$

$$ g(x_1, v_1, x_2) = \frac{dv_1}{dt} = -x_1(k_1 + k_2) -b_1 v_1 + k_2 x_2 + (k_1 - k_2) $$

$$ h(v_2) = \frac{dx_2}{dt} = v_2 $$

$$ i(x_1, x_2, v_2) = \frac{dv_2}{dt} =   k_2 x_1 - k_2 x_2 -b_2 v_2 + k_2 $$

Escribimos la derivada de $\vec{p}$ en términos de una matriz:

$$  \frac{d \vec{p}}{dt} = A\vec{p} + \vec{F} =\begin{bmatrix} 0 & 1 & 0 & 0 \\ -(k_1 + k_2) & -b_1 & k_2 & 0 \\ 0 & 0 & 0 & 1 \\ k_2 & 0 & -k_2 & -b_2 \end{bmatrix} \vec{p} + \begin{pmatrix} 0 \\ k_1 - k_2 \\ 0 \\ k_2 \end{pmatrix}$$ 


Primero buscamos el punto crítico del sistema de ecuaciones no lineal

$$ \frac{d \vec{p}_0}{dt} = A \vec{p}_0 + \vec{F} = 0$$

$$ \vec{p}_0 = -A^{-1} \vec{F} $$

Calculamos la matriz inversa de A:

$$ A^{-1} = \begin{bmatrix} -\frac{b_1}{k_1} & -\frac{1}{k_1} & -\frac{b_2}{k_1} & -\frac{1}{k_1} \\ 1 & 0 & 0 & 0 \\ -\frac{b_1}{k_1} & -\frac{1}{k_1} & -\frac{b_2(k_1 + k_2)}{k_1 k_2} & -\frac{k_1 + k_2}{k_1 k_2} \\ 0 & 0 & 1 & 0 \end{bmatrix} $$ 

$$  $$

Ahora calculamos $\vec{p}_0$:

$$ \vec{p}_0 = -\begin{bmatrix} -\frac{b_1}{k_1} & -\frac{1}{k_1} & -\frac{b_2}{k_1} & -\frac{1}{k_1} \\ 1 & 0 & 0 & 0 \\ -\frac{b_1}{k_1} & -\frac{1}{k_1} & -\frac{b_2(k_1 + k_2)}{k_1 k_2} & -\frac{k_1 + k_2}{k_1 k_2} \\ 0 & 0 & 1 & 0 \end{bmatrix} \begin{pmatrix} 0 \\ k_1 - k_2 \\ 0 \\ k_2 \end{pmatrix} $$

$$ \vec{p}_0 = \begin{pmatrix} 1 \\ 0 \\ 2 \\ 0 \end{pmatrix}$$

Ahora calculamos las derivadas parciales para encontrar ${\bf{J}}$:

$$ \frac{\partial g}{\partial x_1} = -(k_1 + k_2) $$

$$ \frac{\partial g}{\partial v_1} = -b_1 $$

$$ \frac{\partial g}{\partial x_2} = k_2 $$

$$ \frac{\partial i}{\partial x_1} = k_2 $$

$$ \frac{\partial i}{\partial x_2} = -k_2 $$

$$ \frac{\partial i}{\partial v_2} = -b_2 $$

$${\bf{J}} = \begin{bmatrix}
0 & 1 & 0 & 0 \\
-(k_1 + k_2) & -b_1 &
k_2 & 0 \\
0 & 0 & 0 & 1 \\
k_2 & 0 & -k_2 & -b_2 \\
\end{bmatrix} $$ \\

Para conocer la naturaleza del punto crítico es necesario encontrar los eigenvalores de la Matriz Jacobiana. 

### **Caso Sin Fricción**

```python
# Definimos los parámetros:
k1 = 6
k2 = 4
b1 = 0
b2 = 0
```

```python
# Naturaleza del punto crítico:

# Definimos a la matriz A:
A = np.array([[0, 1, 0, 0],[-(k1+k2), -b1, k2, 0], [0,0,0,1], [k2, 0, -k2, -b2]])

# Calculamos los eigenvalores de la matriz:
eigenWert, eigenVektor = la.eig(A)
#print('Los eigenvectores son:')
#print(eigenVektor)
print('Los eigenvalores son:', eigenWert)
```

Los eigenvalores tiene parte real $\approx$ 0. Por lo tanto, no podemos decir nada respecto al punto crítico.

```python
# Definimos el tiempo de integración de 0 a 10 segundos 
t = np.linspace(0, 25, 5000)

# Usamos el método de integración de scipy.integrate.odeint
p = odeint(EDO10, p0, t, args=(k1, k2, b1, b2))


plt.figure(figsize=(8,10))
plt.plot(t, p[:, 0], 'b', label='x1(t)')
plt.plot(t, p[:, 2], 'g', label='x2(t)')
plt.legend(loc='best')
plt.title('Soluciones de x1(t), x2(t) con respecto al tiempo')
plt.xlabel('t')
plt.ylabel(' ')
plt.grid()
plt.show()

plt.figure(figsize=(8,10))
plt.plot(p[:, 0], p[:, 2], 'b')
plt.legend(loc='best')
plt.title('Soluciones de x1(t), x2(t) Espacio Fase')
plt.xlabel('x1(t)')
plt.ylabel('x2(t)')
plt.grid()
plt.show()
```
